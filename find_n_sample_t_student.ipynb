```python
import numpy as np
import scipy.stats as stats

def find_n_sample_t_student(alpha1, alpha2, n_max, range_alpha=0.05, range_n=1):

    alpha_values = np.arange(alpha1, alpha2, range_alpha)
    n_values = np.arange(2, n_max, range_n)
    alphn = np.zeros((len(alpha_values), len(n_values) + 1))
    first_n_dict = {}  # To store the first n values

    for i in range(len(alpha_values)):
        for j in range(len(n_values)):
            alphn[i, j] = 1 - stats.t.cdf(stats.norm.ppf(1 - alpha_values[i]), n_values[j] - 1)


        alphn[:, len(n_values)] = alpha_values
    # alphn = np.round(alphn, 2)

    return  alphn

# Example usage
alpha1 = 0.05
alpha2 = 0.25
n_max = 100

alphn = find_n_sample_t_student(alpha1, alpha2, n_max, range_alpha=0.05, range_n=1)
alphn
```
