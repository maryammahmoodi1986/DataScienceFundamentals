```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from scipy import stats


# --- Min-Max Normalization ---
# Create a copy of the original DataFrame
df_minmax = df.copy()

# Apply MinMaxScaler to 'x1', 'x2', 'x3'
scaler = MinMaxScaler()
df_minmax[['x1_minmax', 'x2_minmax', 'x3_minmax']] = scaler.fit_transform(df_minmax[['x1', 'x2', 'x3']])


# --- Z-score Normalization (Zitski) ---
df_zscore = df.copy()
for col in ['x1', 'x2', 'x3']:
    df_zscore[col + '_zscore'] = (df_zscore[col] - df_zscore[col].mean()) / df_zscore[col].std()


# --- Logarithmic Transformation ---
df_log = df.copy()
for col in ['x1', 'x2', 'x3']:
    # Adding 1 to avoid log(0) issues
    df_log[col + '_log'] = np.log1p(df_log[col])


# --- Square Root Transformation ---
df_sqrt = df.copy()
for col in ['x1', 'x2', 'x3']:
    df_sqrt[col + '_sqrt'] = np.sqrt(df_log[col])


# --- Box-Cox Transformation ---
df_boxcox = df.copy()
for col in ['x1', 'x2', 'x3']:
    # Add a small constant to handle zeros or negative values
    df_boxcox[col + '_boxcox'], _ = stats.boxcox(df_boxcox[col] + 1e-6)


# --- Yeo-Johnson Transformation ---
df_yeojohnson = df.copy()
for col in ['x1', 'x2', 'x3']:
    df_yeojohnson[col + '_yeojohnson'], _ = stats.yeojohnson(df_yeojohnson[col])


# --- Plotting ---
def plot_distributions(data, title, cols):
    fig, axes = plt.subplots(1, 3, figsize=(10, 2))
    fig.suptitle(title, fontsize=16)
    for i, col in enumerate(cols):
        sns.histplot(data[col], ax=axes[i], kde=True)
        axes[i].set_title(col)
    plt.tight_layout()
    plt.show()


plot_distributions(df_minmax, 'Min-Max Normalization', ['x1_minmax', 'x2_minmax', 'x3_minmax'])
plot_distributions(df_zscore, 'Z-score Normalization', ['x1_zscore', 'x2_zscore', 'x3_zscore'])
plot_distributions(df_log, 'Logarithmic Transformation', ['x1_log', 'x2_log', 'x3_log'])
plot_distributions(df_sqrt, 'Square Root Transformation', ['x1_sqrt', 'x2_sqrt', 'x3_sqrt'])
plot_distributions(df_boxcox, 'Box-Cox Transformation', ['x1_boxcox', 'x2_boxcox', 'x3_boxcox'])
plot_distributions(df_yeojohnson, 'Yeo-Johnson Transformation', ['x1_yeojohnson', 'x2_yeojohnson', 'x3_yeojohnson'])
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# لیست ستون‌های دسته‌بندی
categorical_cols = ['shift', 'type', 'obs']

# ایجاد یک ماتریس خالی برای ذخیره نتایج وابستگی
dependency_matrix = pd.DataFrame(index=categorical_cols, columns=categorical_cols, dtype=float) # Initialize with float type

# اجرای آزمون Chi-Square برای هر جفت ستون
for i in categorical_cols:
    for j in categorical_cols:
        if i != j:
            contingency_table = pd.crosstab(df[i], df[j])
            chi2, p, dof, expected = chi2_contingency(contingency_table)

            # تعیین وابستگی براساس p-value
            if p < 0.05:
                dependency_matrix.loc[i, j] = 1  # وابستگی معنی‌دار
            else:
                dependency_matrix.loc[i, j] = 0  # وابستگی معنی‌دار نیست

# نمایش heatmap
plt.figure(figsize=(4, 2))
sns.heatmap(dependency_matrix.astype(float), annot=True, cmap="coolwarm", fmt=".0f", #Ensure matrix is float
            xticklabels=categorical_cols, yticklabels=categorical_cols)
plt.title("Dependency Matrix (Chi-Square Test)")
plt.show()
```

مدل رگرسیون لجستیک چندکلاسه

مدل رگرسیون لجستیک برای  K  کلاس به صورت زیر تعریف می‌شود:


P(y = Cₖ | 𝐱) = e⁽𝐱ᵀ 𝑊ₖ) / ∑ⱼ₌ᴷ e^(𝐱ᵀ 𝑊ⱼ)}


که در آن:

•  P(y = Cₖ | 𝐱)  احتمال تعلق ویژگی‌ها به کلاس  k  است.

•  𝐱  وکتور ویژگی‌ها است.

•  𝑊ₖ  وکتور وزن‌ها برای کلاس  k  است.

<div dir="rtl" style="text-align: center;">


برای رگرسیون لجستیک سه کلاسه با ۵ ویژگی، می‌توانیم از روش One-vs-Rest (OvR) استفاده کنیم. در این روش، برای هر کلاس یک مدل رگرسیون لجستیک جداگانه آموزش داده می‌شود. فرض کنید که کلاس‌ها را با  C₁ ،  C₂  و  C₃  نشان می‌دهیم و ویژگی‌ها را با  x₁, x₂, x₃, x₄, x₅  مشخص می‌کنیم.

▎مدل رگرسیون لجستیک

برای هر کلاس  Cₖ  (که  k = 1, 2, 3 )، می‌توانیم از معادله زیر استفاده کنیم:

P(y = Cₖ | 𝐱) = e⁽zₖ) / e^(z₁ + e^(z₂) + e^(z₃))


که در آن:

zₖ = β₀ₖ + β₁ₖ x₁ + β₂ₖ x₂ + β₃ₖ x₃ + β₄ₖ x₄ + β₅ₖ x₅


در اینجا،  β₀ₖ  ضریب آزاد (intercept) و  βᵢₖ  (برای  i = 1, 2, 3, 4, 5 ) ضرایب مربوط به ویژگی‌ها هستند.

▎دستگاه معادلات

با توجه به توضیحات فوق، می‌توانیم دستگاه معادلات زیر را برای سه کلاس بنویسیم:

1. برای کلاس  C₁ :

P(y = C₁ | 𝐱) = e⁽z₁) / e^(z₁ + e^(z₂) + e^(z₃))


2. برای کلاس  C₂ :

P(y = C₂ | 𝐱) = e⁽z₂) / e^(z₁ + e^(z₂) + e^(z₃))


3. برای کلاس  C₃ :

P(y = C₃ | 𝐱) = e⁽z₃) / e^(z₁ + e^(z₂) + e^(z₃))


▎نتیجه‌گیری

در نهایت، برای هر کلاس، یک مدل رگرسیون لجستیک جداگانه داریم که شامل پارامترهای  β₀ₖ, β₁ₖ, β₂ₖ, β₃ₖ, β₄ₖ, β₅ₖ  است. این ضرایب از طریق الگوریتم‌های بهینه‌سازی آموزش داده می‌شوند.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# فرض بر این است که df از قبل تعریف شده است
X = df[['x1', 'x2', 'x3', 'shift', 'type']]
Y = df['obs']

# اضافه کردن یک ستون برای بایاس
X_biased = np.hstack([np.ones((X.shape[0], 1)), X])  # اضافه کردن بایاس

# محاسبه ضرایب با استفاده از معکوس شبه
beta_pseudo_inverse = np.linalg.pinv(X_biased.T @ X_biased) @ X_biased.T @ Y

# با استفاده از روش ضرب ماتریسی
beta_matrix_multiplication = np.linalg.inv(X_biased.T @ X_biased) @ X_biased.T @ Y

# تجزیه مقدار منفرد
U, S, Vt = np.linalg.svd(X_biased, full_matrices=False)

# محاسبه ضرایب با استفاده از SVD
S_inv = np.diag(1 / S)
beta_svd = Vt.T @ S_inv @ U.T @ Y

# ایجاد DataFrame برای نمایش ضرایب
coefficients_df = pd.DataFrame({
    'Method 1 (Pseudo-inverse)': beta_pseudo_inverse,
    'Method 2 (Matrix Multiplication)': beta_matrix_multiplication,
    'Method 3 (SVD)': beta_svd,
})
coefficients_df
```

```python
import numpy as np
import pandas as pd

# Assuming df is already defined and your previous code to calculate coefficients...

def calculate_k(A):
    """Calculates the condition number of a matrix."""
    # Ensure A is a square matrix before calculating eigenvalues.
    # For SVD method, use A.T @ A to make it square
    if A.shape[0] != A.shape[1]:
        A = A.T @ A  # Make A square if it's not
    eigenvalues, _ = np.linalg.eig(A)
    sigma_max = np.max(np.abs(eigenvalues))
    sigma_min = np.min(np.abs(eigenvalues))
    k = sigma_max / sigma_min
    return k

# Calculate k for each method
A_pseudo_inverse = X_biased.T @ X_biased  # Matrix for Pseudo-inverse method
A_matrix_mult = X_biased.T @ X_biased  # Matrix for Matrix Multiplication method
A_svd = X_biased  # Matrix for SVD method

k_pseudo_inverse = calculate_k(A_pseudo_inverse)
k_matrix_mult = calculate_k(A_matrix_mult)
k_svd = calculate_k(A_svd)  # This will now use the modified calculate_k

print("Condition Number for Method 1 (Pseudo-inverse):", k_pseudo_inverse)
print("Condition Number for Method 2 (Matrix Multiplication):", k_matrix_mult)
print("Condition Number for Method 3 (SVD):", k_svd)
```

مراحل الگوریتم نیوتن-رافسون برای رگرسیون لجستیک چندکلاسه

1. ورودی‌ها:

   • X: ماتریس ویژگی‌ها (N x D) که در آن N تعداد نمونه‌ها و D تعداد ویژگی‌ها است.

   • y: متغیر هدف (N x 1) که شامل برچسب‌های کلاس برای هر نمونه است.

   • num_classes: تعداد کلاس‌های موجود در داده‌ها.

   • max_iter: حداکثر تعداد تکرارها برای همگرایی.

   • tol: آستانه همگرایی.

2. مقداردهی اولیه:

   • پارامترها (W) با مقادیر صفر مقداردهی می‌شوند. W یک ماتریس (D x num_classes) است.

3. حلقه تکرار:

   • برای حداکثر max_iter بار، مراحل زیر انجام می‌شود:

     a. محاسبه گرادیان: با استفاده از تابع gradient، گرادیان تابع هزینه محاسبه می‌شود.

     b. محاسبه هسین ماتریس: با استفاده از تابع hessian، هسین ماتریس محاسبه می‌شود.

     c. تنظیم هسین ماتریس: یک مقدار کوچک به قطر هسین ماتریس اضافه می‌شود تا این ماتریس قابل معکوس شدن باشد (برای جلوگیری از مشکلات عددی).

     d. به‌روزرسانی پارامترها: پارامترهای W با استفاده از فرمول به‌روزرسانی نیوتن-رافسون به‌روزرسانی می‌شوند:

W_(new) = W - H⁻¹ ⋅ ∇ J(W)

که در آن H هسین ماتریس و ∇ J(W) گرادیان تابع هزینه است
      e. بررسی همگرایی: اگر تغییرات پارامترها کمتر از استانه تعیین شده باشد حلقه متوقف می شود.

4. خروجی:

    • پس از اتمام تکرارها یا همگرایی، پارامترهای تخمین زده شده (W) برگردانده می‌شوند.

        hess += np.eye(hess.shape[0]) * 1e-6  
        
        # Update W
        W_new = W - np.linalg.solve(hess, grad)

        # Check for convergence
        if np.linalg.norm(W_new - W) < tol:
            break

        W = W_new
    
    return W


▎توضیحات اضافی:

• Regularization: افزودن یک مقدار کوچک به قطر هسین ماتریس به منظور جلوگیری از مشکلات عددی و اطمینان از معکوس‌پذیری هسین ماتریس انجام می‌شود.

• One-Hot Encoding: برای محاسبه تابع هزینه و گرادیان، برچسب‌ها به صورت one-hot کدگذاری می‌شوند.

```python

# Extract features and target variables
X = df[['x1', 'x2', 'x3','shift','type']].values
y = df[['obs']].values  # Using 'obs' as target

# Number of classes
num_classes = len(np.unique(y))

import numpy as np

def softmax(z):
    """Compute softmax values for each sets of scores in z."""
    e_z = np.exp(z - np.max(z, axis=1, keepdims=True))  # Avoid overflow
    return e_z / np.sum(e_z, axis=1, keepdims=True)

def cost_function(W, X, y, num_classes):
    """Calculate the cost function for multinomial logistic regression."""
    N = X.shape[0]
    scores = X.dot(W)  # Predictions
    probs = softmax(scores)  # Probabilities

    # Cost calculation with one-hot encoding
    y_onehot = np.zeros((N, num_classes))
    y_onehot[np.arange(N), y] = 1

    cost = - (1 / N) * np.sum(y_onehot * np.log(probs))
    return cost

def gradient(W, X, y, num_classes):
    """Calculate the gradient of the cost function."""
    N = X.shape[0]
    scores = X.dot(W)
    probs = softmax(scores)

    # Gradient calculation with one-hot encoding
    y_onehot = np.zeros((N, num_classes))
    y_onehot[np.arange(N), y] = 1

    grad = - (1 / N) * X.T.dot(y_onehot - probs)
    return grad

def hessian(W, X, num_classes):
    """Calculate the Hessian matrix of the cost function."""
    N = X.shape[0]
    scores = X.dot(W)
    probs = softmax(scores)

    hess = np.zeros((W.shape[0], W.shape[0]))
    for i in range(num_classes):
        for j in range(num_classes):
            hess += (1 / N) * X.T.dot(np.diag(probs[:, i] * (int(i == j) - probs[:, j]))).dot(X)

    return hess




def newton_raphson(X, y, num_classes, max_iter=100, tol=1e-6):
    """
    Newton-Raphson method for multinomial logistic regression.

    Parameters:
    X: Feature matrix (N x D)
    y: Target variable (N x 1)
    num_classes: Number of classes
    max_iter: Maximum number of iterations
    tol: Tolerance for convergence

    Returns:
    W: Estimated parameters (D x num_classes)
    """

    # Initialize W with zeros
    N, D = X.shape
    W = np.zeros((D, num_classes))

    for i in range(max_iter):
        grad = gradient(W, X, y, num_classes)
        hess = hessian(W, X, num_classes)

        # Add a small regularization term to the Hessian to make it invertible
        hess = hess + np.eye(hess.shape[0]) * 1e-6

        # Update W using Newton-Raphson update rule
        W_new = W - np.linalg.solve(hess, grad)

        # Check for convergence
        if np.linalg.norm(W_new - W) < tol:
            break

        W = W_new

    return W
# Run the algorithm
W_estimated = newton_raphson(X, y.astype(int) - 1, num_classes)  # Adjusting y for 0-based indexing

print("Estimated Parameters:\n", W_estimated)
```
